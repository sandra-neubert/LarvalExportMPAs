```python
import os
import numpy as np
import xarray as xr
import geopandas as gp
import pandas as pd
```

```python
file = 'OfESncep01globalmmeanu20152019MS.nc'

path = 'C:\\Users\\sandr\\Documents\\Github\\ThesisSandra\\Analysis\\Movement\\TracerDataAndOutput\\OFES\\'
ds = xr.open_dataset(path+file)
ds
```

```python
df = ds.to_dataframe().reset_index(drop=False) 
df
```

```python
dfTest = df[df.latitude == 74.9]
```

```python
dfTest[dfTest.uvel.notna()]
```

```python
df["lon"] = df.longitude
df["lat"] = df.latitude
df = df[df.uvel.notna()]
df = df[df.time == "2015-01-15"]
#df = df[['lon', 'lat']]
df
```

```python
#df.to_csv('C:\\Users\\sandr\\Documents\\Github\\ThesisSandra\\Analysis\\Movement\\Data\\dfOFESOcean.csv')
```

```python
dfBOATS = pd.read_csv('C:\\Users\\sandr\\Documents\\Github\\ThesisSandra\\Analysis\\Movement\\Data\\oceanCellsBOATS.csv', header=None)
dfBOATS = dfBOATS.rename(columns={dfBOATS.columns[0]: 'lon', dfBOATS.columns[1]: 'lat'}).drop(dfBOATS.columns[2], axis=1)
dfBOATS.lon = dfBOATS.lon+180
dfBOATS = dfBOATS[(dfBOATS.lat < 74)&(dfBOATS.lat >(-75))].reset_index(drop=True)
dfBOATS
```

```python
df_merged = dfBOATS.merge(df, how="left", left_on=["lon","lat"], right_on=["lon","lat"], indicator=True)
onlyBOATS = df_merged.query("_merge == 'left_only'")[["lon","lat"]]
```
