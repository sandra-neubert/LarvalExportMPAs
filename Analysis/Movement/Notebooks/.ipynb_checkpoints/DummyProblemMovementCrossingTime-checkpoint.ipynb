## Background

The overall goal is to inlcude larval movement in the ecological module of BOATS. To this end, we explore an appraoch based on van Sebille (2011; https://doi.org/10.1175/2011JPO4602.1) that uses drifter data (https://www.aoml.noaa.gov/phod/gdp/hourly_data.php) together with OFES data to create a transition matrix giving the probability of moving from a $cell_{i,j}$ to the four directly adjacent cells. The probabilities are associated with the average time it takes for the crossing to happen, hereby representing a proxy for the speed of the crossing. These crossing times can then be used to downscale the probabilities to bring them all to the same time scale.

## Algorithm

1. Pre-processing of the drifter data.
2. For each cell, <br/>
    2.1. extract how many drifters have crossed this cell at some point over the time scale of the dataset (here: $TotalDrifters_{i,j} = 10$) and the indices of the neighbouring cells. <br/>
    2.2. sum how many of the drifters crossing the cell left into a specific direction (North, South, West or East; e.g. $DrifterCrossings_{{i,j}_{West}} = 3$). Each crossing is associated with a specific time scale (duration of the crossing), which can be averaged across all drifters going towards a specific direction (e.g. $AvgCrossingTime_{{i,j}_{West}} = 50 h$). <br/>
    2.3. calculate the probability that a drifter leaves the cell into a specific direction: <br/> $Pr(DrifterCrossings_{{i,j}_{West}}) = DrifterCrossings_{{i,j}_{West}} / TotalDrifters_{i,j} = 3/10$ <br/>
    2.4. "normalize" the calculated probability with time to have a common time scale which can be used when integrating the movement into the ecological model. We use the minimum average crossing time here, but any time smaller than the average crossing time can be used for normalization: <br/> 
   $Pr_{Norm}(DrifterCrossings_{{i,j}_{West}}) = Pr(DrifterCrossings_{{i,j}_{West}})  / AvgCrossingTime^{Norm}_{{i,j}_{West}} = 0.3/10,$ where $min(AvgCrossingTime) = 5 h$ and $AvgCrossingTime^{Norm}_{{i,j}_{West}} = AvgCrossingTime_{{i,j}_{West}}/ min(AvgCrossingTime) = 50/5$ 

    *Values are examples and not based on real data.*

### Load packages

```python
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import animation
from IPython.display import HTML

# Enable interactive plot
%matplotlib inline
#%matplotlib notebook
```

### Define functions

```python
def indicesNeighbours(i, j, MatInd, dimM, dimP):
    #inputs: 
    # i: row Number
    # j: col Number
    # MatInd: matrix with indices for biomass matrix cells
    # dimM: x-dimension of biomass matrix
    # dimP: x-dimension of probability matrix

    #outputs:
    # cellNum: cell currently looked at and the four directly
    # adjacent cells: N(orth),S(outh),W(est),E(ast)
    
    cellNum = MatInd[i,j]
    
    if i == 0: #Later: can just set probability to 0 for going up or down from first or last row respectively
        N = MatInd[i,j] + dimM - 1
        S = MatInd[i,j] + 1

    elif i == (dimM-1):
        N = MatInd[i,j] - 1
        S = MatInd[i,j] - dimM + 1

    else:
        N = MatInd[i,j] -1
        S = MatInd[i,j] +1
            
    if j == 0:
        W = MatInd[i,j] + dimP - dimM
        E = MatInd[i,j] + dimM

    elif j == (dimM-1):
        W = MatInd[i,j] - dimM
        E = MatInd[i,j] -dimP + dimM

    else: 
        W = MatInd[i,j] - dimM
        E = MatInd[i,j] + dimM
                
    return cellNum, N, S, W, E

def createCrossM(dimP, dimM, MatInd, NC, SC, WC, EC):
    #inputs: 
    # MatInd: matrix with indices for biomass matrix cells
    # dimM: x-dimension of biomass matrix
    # dimP: x-dimension of probability/crossing matrix
    # NC, SC, WC, EC: average crossing times to move North, South, West, East 

    #outputs:
    # C: Crossing matrix with average crossing times
    # normTime: minimum crossing time (to be used later to run model)
    
    C = np.zeros((dimP,dimP)) #init matrix (dim = max(index) x max(index))

    for i in range(dimM): #loop through rows and columns, get indices of cell i,j and four directly adjacent cells
        for j in range(dimM):
            [cellNum, N, S, W, E] = indicesNeighbours(i, j, MatInd, dimM, dimP) # call function to get index of current cell + 4 adjacent ones
            
            C[cellNum, N] = NC #prob move up
            C[cellNum, S] = SC #down
            C[cellNum, W] = WC #left
            C[cellNum, E] = EC #right
    
    normTime = np.min(C[np.nonzero(C)]) #get minimum crossing time to use as time step in later model
    return(C, normTime)

def createProbM(dimP, dimM, MatInd, NP, SP, WP, EP):
    #inputs: 
    # MatInd: matrix with indices for biomass matrix cells
    # dimM: x-dimension of biomass matrix
    # dimP: x-dimension of probability matrix
    # NP, SP, WP, EP: Probabilities to move North, South, West, East (should all add up to 1!)

    #outputs:
    # P: Probability matrix with transition probabilities
    
    #sanity check 
    if np.round((NP + SP + WP + EP), decimals = 2) != 1:
        print("Probabilities don't add up to 1. Change input.")
        return
    
    P = np.zeros((dimP,dimP)) #init matrix (dim = max(index) x max(index))

    for i in range(dimM): #loop through rows and columns, get indices of cell i,j and four directly adjacent cells
        for j in range(dimM):
            [cellNum, N, S, W, E] = indicesNeighbours(i, j, MatInd, dimM, dimP) # call function to get index of current cell + 4 adjacent ones
            
            P[cellNum, N] = NP #prob move up
            P[cellNum, S] = SP #down
            P[cellNum, W] = WP #left
            P[cellNum, E] = EP #right
            
    return(P)

def createNormProbM(dimP, dimM, MatInd, P, C, normTime):
    #inputs: 
    # MatInd: matrix with indices for biomass matrix cells
    # dimM: x-dimension of biomass matrix
    # dimP: x-dimension of probability matrix
    # P: Probability matrix with transition probabilities
    # C: Crossing matrix with average crossing times
    # normTime: timestep used to normalize probability matrix
    
    #outputs:
    # PN: Probability matrix with transition probabilities normalized with time

    PN = np.zeros((dimP,dimP)) #init matrix (dim = max(index) x max(index))

    for i in range(dimM): #loop through rows and columns, get indices of cell i,j and four directly adjacent cells
        for j in range(dimM):
            [cellNum, N, S, W, E] = indicesNeighbours(i, j, MatInd, dimM, dimP) # call function to get index of current cell + 4 adjacent ones
            neighbours = (N, S, W, E)
            neighProb = []
            for direction in range(len(neighbours)):
                currentNeighb = neighbours[direction]
                NeighCN = C[cellNum, currentNeighb]/normTime #avg crossing time to any direction normalized with time step
                
                if NeighCN == 0: #otherwise might divide by 0
                    NeighPN = 0
                else:
                    NeighPN = np.round((P[cellNum, currentNeighb]/NeighCN), decimals = 2) #time normalised prob to go any direction 
                
                PN[cellNum, currentNeighb] = NeighPN
                neighProb.append(NeighPN)
            
            StayPN = 1 - (np.round(sum(neighProb), decimals = 2)) 
            
            PN[cellNum, cellNum] = StayPN #prob stay in same cell
            
    return(PN)

def movement(t, dimM, dimP, endI, endJ, MatBiomass, MatInd, MatProb, amountToMove):
    #inputs: 
    # t: time (e.g. 100)
    # dimM: x-dimension of biomass matrix
    # dimP: x-dimension of probability matrix
    # endI: end (rows) of cells to be indexed for movement (ONLY FOR SANITY CHECK, REPLACE WITH dimM LATER!!!!)
    # endJ: end (cols) of cells to be indexed for movement (ONLY FOR SANITY CHECK, REPLACE WITH dimM LATER!!!!)
    # MatBiomass: matrix with biomass to be moved
    # MatInd: matrix with indices for biomass matrix cells
    # MatProb: matrix with transition probabilities
    # amountToMove: amount of biomass moved at each time step

    #outputs:
    # biomassToMoveM: 3-D matrix (t, dimM, dimM) with the biomass in each cell at each time step (sum at each t should be same)
   
    biomassToMoveM = np.zeros((t, dimM, dimM)) #init matrix
    biomassToMoveM[0,:,:] = MatBiomass #give starting values

    for time in np.arange(1,t): #loop through time

        biomassToMoveM[time,:,:] = biomassToMoveM[time-1, :,:] #init each new time step with biomass from previous time step (which is then changed)
        #biomassToMoveM = np.stack([biomassToMoveM, matNow])
        for i in range(endI):
            for j in range(endJ):
                [cellNum, N, S, W, E] = indicesNeighbours(i, j, MatInd, dimM, dimP) # call function to get index of current cell + 4 adjacent ones
                Prob = [MatProb[cellNum, cellNum], MatProb[cellNum, N], MatProb[cellNum, S], MatProb[cellNum, W], MatProb[cellNum, E]]
                X = [cellNum, N, S, W, E]
                n = 1
                moveTo = random.choices(X, weights = Prob,  k = n) #draw cell from cell i,j and four neighbours based on probability distribution
                moveTo = moveTo[0] #just to avoid indexing later
            
                move_out = -amountToMove #taken from cell i,j
                move_in = amountToMove
     
                rowCellNum, colCellNum = np.where(MatInd == cellNum) #get location of where biomass moves to
                CellNumValFuture = biomassToMoveM[time, rowCellNum[0], colCellNum[0]] + move_out #to prevent biomass to be negative
            
                if (moveTo == cellNum) | (CellNumValFuture < 0):#(biomassToMoveM[time, i, j] == 0): #either stay or moving would make cell biomass negative
                    biomassToMoveM[time, i, j] = biomassToMoveM[time, i, j] #biomass in cell i,j same as in previous time step
                    #print("stay")
                else:
                    biomassToMoveM[time, i, j] = biomassToMoveM[time, i, j] + move_out #take biomass away
                    rowMove, colMove = np.where(MatInd == moveTo)
                        #M.biomassToMove(rowMove, colMove, t) = M.biomassToMove(rowMove, colMove, t-1) + delta_t*move_in;
                    biomassToMoveM[time, rowMove[0], colMove[0]]  = biomassToMoveM[time, rowMove[0], colMove[0]] + move_in #add biomass
               
            
    print("Start Biomass:", np.sum(biomassToMoveM[0,:,:]), "End Biomass:", np.sum(biomassToMoveM[t-1,:,:]))
    return biomassToMoveM
```

### Initialise matrices

Initialise the biomass matrix and the index matrix needed for the creation of the probability matrix.

```python
biomass=[]
n=100
for i in range(n):
    biomass.append(random.randint(30,100)) #Here: random values of biomass for each cell

biomassM = np.array(biomass).reshape(10,10)
indexCellM = np.arange(0,100).reshape(10,10).transpose() #transpose to make indices same order as in matlab

dimM = np.size(biomassM,0)
dimP = np.size(biomassM)
```

### Create Probability Matrix

Create matrix that gives the probability to get from every cell (indexed) to every other cell. Because transitions are only possible to the four adjacent cells, the matrix is mostly filled with zero values.

```python
P = createProbM(dimP, dimM, MatInd = indexCellM, NP = 0.4, SP = 0, WP = 0.3, EP = 0.3)
```

### Create Crossing Times Matrix

Create matrix that gives the average time it takes to get from every cell to the four directly adjacent cells. The minimum time across the whole matrix will be used as the time steps for the movement.

```python
C, timestep = createCrossM(dimP, dimM, MatInd = indexCellM, NC = 5, SC = 0, WC = 50, EC = 30)
print(timestep)
```

### "Normalize" Probability Matrix using Crossing Times

To allow for the use of the transition probability to move biomass in a model at a specific time step, the different probabilities have to be normalized with time. For this, we used the minimum average crossing time found in the crossing times matrix, which will later be used as the time step for the model. 

```python
PN = createNormProbM(dimP, dimM, MatInd = indexCellM, P = P, C = C, normTime = timestep)
```

### Movement

Run the actual movement. For a specific cell i,j, a movement option is chosen based on the provided probabilities (a neighbouring cell or "stay" option drawn from provided probability distribution). This leads to biomass being subtracted from cell i,j (if option != "stay") and the same amount being added to the drawn adjacent cell to prevent biomass from being created or destroyed. 

```python
biomassToMoveM = movement(t = 100, dimM = dimM, dimP = dimP, endI = dimM, endJ = dimM, 
                          MatBiomass = biomassM, MatInd = indexCellM, MatProb = PN, amountToMove = 10)
```

### Visualise time series

```python
#plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

fig, ax = plt.subplots()
nx = dimM
ny = dimM

maxVal = 100 #to see outcome better

line2d, = ax.plot([], [], lw=2)

ax_global = ax

data = biomassToMoveM[0,:,:]
sns.heatmap(data, vmin=0, vmax=maxVal)

def init_heatmap():
    line2d.set_data([], [])
    sns.heatmap(data, ax=ax_global,  vmin=0, vmax=maxVal, cbar=False)
    return (line2d,)


def animate_heatmap(i):
    data = biomassToMoveM[i,:,:]
    sns.heatmap(data, ax=ax_global, vmin=0, vmax=maxVal, cbar=False)
    ax.set_title('Frame: ' + str(i))    

    return (line2d,)

anim = animation.FuncAnimation(fig, animate_heatmap, init_func=init_heatmap,
                              frames=100, interval=100, blit=True, repeat=False)

plt.close()
HTML(anim.to_html5_video())    
```

Problem: what should Pr(stay)be? With the current algorithm, Pr(stay) = 1 - sum(move any direction), which means if sum(move) > 1, then Pr(stay) < 0. Should Pr(stay) be manually set to 0?

Solution: Use a timestep that's small enough to avoid this problem. Set time use to normalize as the min(AvgCrossingTime).
